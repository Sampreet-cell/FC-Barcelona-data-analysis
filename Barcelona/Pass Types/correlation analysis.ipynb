```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Pass types.csv")
```

```python
attributes = ["Attempted",
              "Live Pass",
              "Dead Pass",
              "Free Kick Pass",
              "Through Pass",
              "Switching Passes",
              "Cross Pass",
              "Throw ins",
              "Corner Kick Pass",
              "CK Inswing",
              "CK Outswing",
              "CK Straight",
              "Cmp",
              "Offside Pass",
              "Blocks"
]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Pass Types")
plt.show()
```
