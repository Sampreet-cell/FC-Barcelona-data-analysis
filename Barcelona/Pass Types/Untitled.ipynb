```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Pass types.csv")
```

```python
player_attributes = data[["Player", "Attempted", "Cmp", "Live Pass", "Dead Pass", "Through Pass", "Cross Pass"]]
```

```python
top_att_players = player_attributes.sort_values(by="Attempted", ascending=False).head(5)
```

```python
top_cmp_players = player_attributes.sort_values(by="Cmp", ascending=False).head(5)
```

```python
top_live_players = player_attributes.sort_values(by="Live Pass", ascending=False).head(5)
```

```python
top_dead_players = player_attributes.sort_values(by="Dead Pass", ascending=False).head(5)
```

```python
top_through_players = player_attributes.sort_values(by="Through Pass", ascending=False).head(5)
```

```python
top_cross_players = player_attributes.sort_values(by="Cross Pass", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(6, 1, figsize=(10, 15))
axs[0].bar(top_att_players["Player"], top_att_players["Attempted"])
axs[0].set_title("Top 5 Players - Passes Attempted")
axs[0].set_ylabel("Attempted")

axs[1].bar(top_cmp_players["Player"], top_cmp_players["Cmp"])
axs[1].set_title("Top 5 Players - Passes Completed")
axs[1].set_ylabel("Completed")

axs[2].bar(top_live_players["Player"], top_live_players["Live Pass"])
axs[2].set_title("Top 5 Players - Live Passes Attempted")
axs[2].set_ylabel("Live Pass")

axs[3].bar(top_dead_players["Player"], top_dead_players["Dead Pass"])
axs[3].set_title("Top 5 Players - Dead Passes Attempted")
axs[3].set_ylabel("Dead Pass")

axs[4].bar(top_through_players["Player"], top_through_players["Through Pass"])
axs[4].set_title("Top 5 Players - Through Passes Attempted")
axs[4].set_ylabel("Through Pass")

axs[5].bar(top_cross_players["Player"], top_cross_players["Cross Pass"])
axs[5].set_title("Top 5 Players - Cross Passes Attempted")
axs[5].set_ylabel("Cross Pass")

plt.tight_layout()
plt.show()
```
