```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Possession barcelona.csv")
```

```python
player_attributes = data[["Player", "Touches ratio", "Carries ratio", "Receiving ratio", "Receiving PrgR ratio", "Carries one third ratio", "CPA ratio", "PrgDist ratio", "Dis ratio"]]
```

```python
top_touches_players = player_attributes.sort_values(by="Touches ratio", ascending=False).head(5)
```

```python
top_carries_players = player_attributes.sort_values(by="Carries ratio", ascending=False).head(5)
```

```python
top_receiving_players = player_attributes.sort_values(by="Receiving ratio", ascending=False).head(5)
```

```python
top_receiving_prg_players = player_attributes.sort_values(by="Receiving PrgR ratio", ascending=False).head(5)
```

```python
top_carries_one_third_players = player_attributes.sort_values(by="Carries one third ratio", ascending=False).head(5)
```

```python
top_CPA_players = player_attributes.sort_values(by="CPA ratio", ascending=False).head(5)
```

```python
top_PrgDist_players = player_attributes.sort_values(by="PrgDist ratio", ascending=False).head(5)
```

```python
top_Dis_players = player_attributes.sort_values(by="Dis ratio", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(8, 1, figsize=(10, 15))
axs[0].bar(top_touches_players["Player"], top_touches_players["Touches ratio"])
axs[0].set_title("Top 5 Players - Touches")
axs[0].set_ylabel("Touches")

axs[1].bar(top_carries_players["Player"], top_carries_players["Carries ratio"])
axs[1].set_title("Top 5 Players - Carries")
axs[1].set_ylabel("Carries")

axs[2].bar(top_receiving_players["Player"], top_receiving_players["Receiving ratio"])
axs[2].set_title("Top 5 Players - Receiving")
axs[2].set_ylabel("Receiving")

axs[3].bar(top_receiving_prg_players["Player"], top_receiving_prg_players["Receiving PrgR ratio"])
axs[3].set_title("Top 5 Players - Progressive Receiving")
axs[3].set_ylabel("Progressive Receiving")

axs[4].bar(top_carries_one_third_players["Player"], top_carries_one_third_players["Carries one third ratio"])
axs[4].set_title("Top 5 Players - Carries 1/3")
axs[4].set_ylabel("Carries 1/3")

axs[5].bar(top_CPA_players["Player"], top_CPA_players["CPA ratio"])
axs[5].set_title("Top 5 Players - Carries to Penalty Area")
axs[5].set_ylabel("Carries to Penalty Area")

axs[6].bar(top_PrgDist_players["Player"], top_PrgDist_players["PrgDist ratio"])
axs[6].set_title("Top 5 Players - Progressive Distance")
axs[6].set_ylabel("Progressive Distance")

axs[7].bar(top_Dis_players["Player"], top_Dis_players["Dis ratio"])
axs[7].set_title("Top 5 Players - Dispossessed")
axs[7].set_ylabel("Dispossessed")

plt.tight_layout()
plt.show()
```
