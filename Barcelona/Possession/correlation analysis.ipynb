```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Possession barcelona.csv")
```

```python
attributes = ["Touches ratio",
              "Def Pen ratio",
              "Def 3rd ratio",
              "Mid 3rd ratio",
              "Att 3rd ratio",
              "Att Pen ratio",
              "Take-Ons Att ratio",
              "Take-Ons Succ ratio",
              "Take-Ons Succ rate ratio",
              "Take-Ons Tkld ratio",
              "Take-Ons Tkld rate ratio",
              "Carries ratio",
              "TotDist ratio",
              "PrgDist ratio",
              "PrgC ratio",
              "Carries one third ratio",
              "CPA ratio",
              "Miscontrols ratio",
              "Dis ratio",
              "Receiving ratio",
              "Receiving PrgR ratio"]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Possession")
plt.show()
```
