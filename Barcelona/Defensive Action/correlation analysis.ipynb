```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("defensive action barcelona.csv")
```

```python
attributes = ["Tackles ratio",
            "Tackles TklW ratio",
            "Tackles Def 3rd ratio",
            "Tackles Mid 3rd ratio",
            "Tackles Att 3rd ratio",
            "Dribblers tackled ratio",
            "Dribbles challenged ratio",
            "Percent of dribblers tackled ratio", 
            "Challenges Lost ratio", 
            "Blocks ratio", 
            "Blocks Sh ratio", 
            "Blocks Pass ratio",
            "Int ratio",
            "Tkl Int ratio",
            "Clr ratio",
            "Err ratio"
]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Defensive Actions")
plt.show()
```
