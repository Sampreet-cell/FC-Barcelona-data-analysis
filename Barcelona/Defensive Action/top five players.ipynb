```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("defensive action barcelona.csv")
```

```python
player_attributes = data[["Player", "Tackles ratio", "Tackles TklW ratio", "Int ratio", "Tkl Int ratio", "Blocks ratio", "Clr ratio", "Err ratio"]]
```

```python
top_tackles_players = player_attributes.sort_values(by="Tackles ratio", ascending=False).head(5)
```

```python
top_tackles_won_players = player_attributes.sort_values(by="Tackles TklW ratio", ascending=False).head(5)
```

```python
top_interceptions_players = player_attributes.sort_values(by="Int ratio", ascending=False).head(5)
```

```python
top_blocks_players = player_attributes.sort_values(by="Blocks ratio", ascending=False).head(5)
```

```python
top_clearances_players = player_attributes.sort_values(by="Clr ratio", ascending=False).head(5)
```

```python
top_error_players = player_attributes.sort_values(by="Err ratio", ascending=False).head(5)
```

```python
top_tackles_interceptions_players = player_attributes.sort_values(by="Tkl Int ratio", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(7, 1, figsize=(10, 15))
axs[0].bar(top_tackles_won_players["Player"], top_tackles_won_players["Tackles TklW ratio"])
axs[0].set_title("Top 5 Players - Tackles Won")
axs[0].set_ylabel("Tackles Won")

axs[1].bar(top_interceptions_players["Player"], top_interceptions_players["Int ratio"])
axs[1].set_title("Top 5 Players - Interceptions")
axs[1].set_ylabel("Interceptions")

axs[2].bar(top_blocks_players["Player"], top_blocks_players["Blocks ratio"])
axs[2].set_title("Top 5 Players - Blocks")
axs[2].set_ylabel("Blocks")

axs[3].bar(top_clearances_players["Player"], top_clearances_players["Clr ratio"])
axs[3].set_title("Top 5 Players - Clearances")
axs[3].set_ylabel("Clearances")

axs[4].bar(top_error_players["Player"], top_error_players["Err ratio"])
axs[4].set_title("Top 5 Players - Errors")
axs[4].set_ylabel("Errors")

axs[5].bar(top_tackles_interceptions_players["Player"], top_tackles_interceptions_players["Tkl Int ratio"])
axs[5].set_title("Top 5 Players - Tackles + Interceptions")
axs[5].set_ylabel("Tackles + Interceptions")

axs[6].bar(top_tackles_players["Player"], top_tackles_players["Tackles ratio"])
axs[6].set_title("Top 5 Players - Tackles")
axs[6].set_ylabel("Tackles")

plt.tight_layout()
plt.show()
```
