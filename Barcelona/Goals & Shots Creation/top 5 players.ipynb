```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Goals shots creation.csv")
```

```python
player_attributes = data[["Player", "SCA ratio", "GCA ratio", "GCA PassLive ratio", "SCA PassLive ratio", "GCA Sh ratio"]]
```

```python
top_SCA_players = player_attributes.sort_values(by="SCA ratio", ascending=False).head(5)
```

```python
top_GCA_players = player_attributes.sort_values(by="GCA ratio", ascending=False).head(5)
```

```python
top_GCA_live_players = player_attributes.sort_values(by="GCA PassLive ratio", ascending=False).head(5)
```

```python
top_SCA_live_players = player_attributes.sort_values(by="SCA PassLive ratio", ascending=False).head(5)
```

```python
top_GCA_shoot_players = player_attributes.sort_values(by="GCA Sh ratio", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10, 15))
axs[0].bar(top_SCA_players["Player"], top_SCA_players["SCA ratio"])
axs[0].set_title("Top 5 Players - Shots creation action")
axs[0].set_ylabel("Shots creation")

axs[1].bar(top_GCA_players["Player"], top_GCA_players["GCA ratio"])
axs[1].set_title("Top 5 Players - Goals creation action")
axs[1].set_ylabel("Goals creation")

axs[2].bar(top_GCA_live_players["Player"], top_GCA_live_players["GCA PassLive ratio"])
axs[2].set_title("Top 5 Players - Goals creation from live pass")
axs[2].set_ylabel("GCA live pass")

axs[3].bar(top_SCA_live_players["Player"], top_SCA_live_players["SCA PassLive ratio"])
axs[3].set_title("Top 5 Players - Shots creation from live pass")
axs[3].set_ylabel("SCA live pass")

axs[4].bar(top_GCA_shoot_players["Player"], top_GCA_shoot_players["GCA Sh ratio"])
axs[4].set_title("Top 5 Players - Goals creation from shots")
axs[4].set_ylabel("GCA shots")

plt.tight_layout()
plt.show()
```
