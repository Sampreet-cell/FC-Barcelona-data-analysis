```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Goals shots creation.csv")
```

```python
attributes = ["SCA ratio",
              "SCA PassLive ratio",
              "SCA PassDead ratio",
              "SCA TO ratio",
              "SCA Sh ratio",
              "SCA Fld ratio",
              "SCA Def ratio",
              "GCA ratio",
              "GCA PassLive ratio",
              "GCA PassDead ratio",
              "GCA TO ratio",
              "GCA Sh ratio",
              "GCA Fld ratio",
              "GCA Def ratio"
]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Goals & Shots creation")
plt.show()
```
