```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("shooting.csv")
```

```python
player_attributes = data[["Player", "Goals", "Shot", "SoT", "Goals by Shot", "Goals by Sot", "Avg Distance"]]
```

```python
top_goal_players = player_attributes.sort_values(by="Goals", ascending=False).head(5)
```

```python
top_shot_players = player_attributes.sort_values(by="Shot", ascending=False).head(5)
```

```python
top_sot_players = player_attributes.sort_values(by="SoT", ascending=False).head(5)
```

```python
top_gbs_players = player_attributes.sort_values(by="Goals by Shot", ascending=False).head(5)
```

```python
top_gbsot_players = player_attributes.sort_values(by="Goals by Sot", ascending=False).head(5)
```

```python
top_dist_players = player_attributes.sort_values(by="Avg Distance", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(6, 1, figsize=(10, 15))
axs[0].bar(top_goal_players["Player"], top_goal_players["Goals"])
axs[0].set_title("Top 5 Players - Goals scored")
axs[0].set_ylabel("Goals")

axs[1].bar(top_shot_players["Player"], top_shot_players["Shot"])
axs[1].set_title("Top 5 Players - Shot attempted")
axs[1].set_ylabel("Shots")

axs[2].bar(top_sot_players["Player"], top_sot_players["SoT"])
axs[2].set_title("Top 5 Players - Shots on Target")
axs[2].set_ylabel("Shots on Target")

axs[3].bar(top_gbs_players["Player"], top_gbs_players["Goals by Shot"])
axs[3].set_title("Top 5 Players - Goals/Shot")
axs[3].set_ylabel("Goals by Shots")

axs[4].bar(top_gbsot_players["Player"], top_gbsot_players["Goals by Sot"])
axs[4].set_title("Top 5 Players - Goals/SoT")
axs[4].set_ylabel("Goals by SoT")

axs[5].bar(top_dist_players["Player"], top_dist_players["Avg Distance"])
axs[5].set_title("Top 5 Players - Average shot distance")
axs[5].set_ylabel("Distance")

plt.tight_layout()
plt.show()
```
