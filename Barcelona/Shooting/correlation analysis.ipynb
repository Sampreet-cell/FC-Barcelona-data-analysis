```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("shooting.csv")
```

```python
attributes = ["Goals",
              "Shot",
              "SoT",
              "Goals by Shot",
              "Goals by Sot",
              "Avg Distance",
              "FK",
              "PK",
              "PK att"
]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Shooting")
plt.show()
```
