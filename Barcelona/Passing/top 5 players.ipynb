```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Passing.csv")
```

```python
player_attributes = data[["Player", "Total Cmp", "Total Att", "Ast", "Total TotDist", "KP", "PrgP"]]
```

```python
top_cmp_players = player_attributes.sort_values(by="Total Cmp", ascending=False).head(5)
```

```python
top_att_players = player_attributes.sort_values(by="Total Att", ascending=False).head(5)
```

```python
top_ast_players = player_attributes.sort_values(by="Ast", ascending=False).head(5)
```

```python
top_dist_players = player_attributes.sort_values(by="Total TotDist", ascending=False).head(5)
```

```python
top_kp_players = player_attributes.sort_values(by="KP", ascending=False).head(5)
```

```python
top_prg_players = player_attributes.sort_values(by="PrgP", ascending=False).head(5)
```

```python
fig, axs = plt.subplots(6, 1, figsize=(10, 15))
axs[0].bar(top_cmp_players["Player"], top_cmp_players["Total Cmp"])
axs[0].set_title("Top 5 Players - Total Passes Completed")
axs[0].set_ylabel("Total pass completion")

axs[1].bar(top_att_players["Player"], top_att_players["Total Att"])
axs[1].set_title("Top 5 Players - Total Passes Attempted")
axs[1].set_ylabel("Total pass attempted")

axs[2].bar(top_ast_players["Player"], top_ast_players["Ast"])
axs[2].set_title("Top 5 Players - Assists")
axs[2].set_ylabel("Assists")

axs[3].bar(top_dist_players["Player"], top_dist_players["Total TotDist"])
axs[3].set_title("Top 5 Players - Total Passing Distance")
axs[3].set_ylabel("Total passing distance")

axs[4].bar(top_kp_players["Player"], top_kp_players["KP"])
axs[4].set_title("Top 5 Players - Key Passes")
axs[4].set_ylabel("Key passes")

axs[5].bar(top_prg_players["Player"], top_prg_players["PrgP"])
axs[5].set_title("Top 5 Players - Progressive Passes")
axs[5].set_ylabel("Progressive passes")

plt.tight_layout()
plt.show()
```
