```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Passing.csv")
```

```python
attributes = ["Total Cmp",
              "Total Att",
              "Total TotDist",
              "Total PrgDist",
              "Short Cmp",
              "Short Att",
              "Medium Cmp",
              "Medium Att",
              "Long Cmp",
              "Long Att",
              "Ast",
              "KP",
              "one third",
              "PPA",
              "CrsPA",
              "PrgP"
]
```

```python
correlation_matrix = data[attributes].corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Passing")
plt.show()
```
